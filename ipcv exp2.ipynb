#Digital Negative

```python
def apply_digital_negative(image):
    negative_image = 255 - image
    return negative_image

num_images_to_display = 9

for i in range(num_images_to_display):

    negative_image = apply_digital_negative(train_X[i])
    plt.subplot(3, 3, i + 1 )
    plt.imshow(negative_image, cmap='gray')

plt.show()
```

#Threshold

```python
def apply_threshold(image, threshold):
    thresholded_image = np.where(image > threshold, 255, 0)
    return thresholded_image

num_images_to_display = 9
binary_threshold = 128

plt.figure(figsize=(12, 12))
for i in range(num_images_to_display):
    plt.subplot(3, 3, i + 1)
    original_image = train_X[i]
    thresholded_image = apply_threshold(original_image, binary_threshold)

    plt.imshow(thresholded_image, cmap='gray')
    plt.title(f'Image {i+1}')

plt.show()
```

#Gray level slicing


```python
def gray_level_slicing(image, lower_threshold, upper_threshold):
    sliced_image = np.where((image >= lower_threshold) & (image <= upper_threshold), image, 0)
    return sliced_image

num_images_to_display = 9
lower_threshold = 100
upper_threshold = 200

plt.figure(figsize=(12, 12))
for i in range(num_images_to_display):
    plt.subplot(3, 3, i + 1)
    original_image = train_X[i]
    sliced_image = gray_level_slicing(original_image, lower_threshold, upper_threshold)

    plt.imshow(sliced_image, cmap='gray')
    plt.title(f'Image {i+1}')

plt.show()
```
