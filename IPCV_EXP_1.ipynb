```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mnist = tf.keras.datasets.mnist
(train_images, train_labels), (_, _) = mnist.load_data()

def get_random_images_by_digit(digit, num_images=1):
    digit_indices = np.where(train_labels == digit)[0]
    random_indices = np.random.choice(digit_indices, num_images, replace=False)
    return train_images[random_indices]
digit = 7
num_images = 5

random_images = get_random_images_by_digit(digit, num_images)

for image in random_images:
    plt.imshow(image, cmap='gray')
    plt.title(f"Digit: {digit}")
    plt.axis('off')
    plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Load the MNIST dataset
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (_, _) = mnist.load_data()

# Select a random image from the dataset
random_index = np.random.randint(0, len(train_images))
original_image = train_images[random_index]

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

# Display the grayscale image
plt.subplot(1, 2, 2)
plt.imshow(original_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

```python
#Task 4

# Define the coordinates for cropping (top-left and bottom-right)
crop_top, crop_left = 7, 7
crop_bottom, crop_right = 21, 21

# Crop the original image
cropped_image = original_image[crop_top:crop_bottom, crop_left:crop_right]

plt.subplot(1, 2, 2)
plt.imshow(cropped_image, cmap='gray')
plt.title('Cropped Image')
plt.axis('off')
```

```python
#Task 5, 6, 7, 8, 9

import cv2
# Select two random images from the dataset
index1, index2 = np.random.randint(0, len(train_images), size=2)
image1, image2 = train_images[index1], train_images[index2]

# Perform arithmetic operations on the images
addition_result = cv2.add(image1, image2)
subtraction_result = cv2.subtract(image1, image2)
multiplication_result = cv2.multiply(image1, image2)
division_result = cv2.divide(image1.astype(float), image2.astype(float))
blending_result = (image1 + image2) // 2  # Simple blending by averaging pixel values

# Display the original images and results of arithmetic operations
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(image1, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(image2, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(addition_result, cmap='gray')
plt.title('Addition')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(subtraction_result, cmap='gray')
plt.title('Subtraction')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(multiplication_result, cmap='gray')
plt.title('Multiplication')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(division_result, cmap='gray')
plt.title('Division')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

index1, index2 = np.random.randint(0, len(train_images), size=2)
image1, image2 = train_images[index1], train_images[index2]

_, image1_binary = cv2.threshold(image1, 127, 255, cv2.THRESH_BINARY)
_, image2_binary = cv2.threshold(image2, 127, 255, cv2.THRESH_BINARY)

and_result = cv2.bitwise_and(image1_binary, image2_binary)
or_result = cv2.bitwise_or(image1_binary, image2_binary)
not_result1 = cv2.bitwise_not(image1_binary)
not_result2 = cv2.bitwise_not(image2_binary)
xor_result = cv2.bitwise_xor(image1_binary, image2_binary)

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(image1, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(image2, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(and_result, cmap='gray')
plt.title('AND')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(or_result, cmap='gray')
plt.title('OR')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(xor_result, cmap='gray')
plt.title('XOR')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(not_result1, cmap='gray')
plt.title('NOT of Image 1')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#Task 14, 15


index = np.random.randint(0, len(train_images))
image = train_images[index]

image = np.uint8(image)

left_shift_result = cv2.bitwise_and(cv2.bitwise_and(image << 2, 255), 255)
right_shift_result = cv2.bitwise_and(cv2.bitwise_and(image >> 2, 255), 255)

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(left_shift_result, cmap='gray')
plt.title('Left Shift (<< 2)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(right_shift_result, cmap='gray')
plt.title('Right Shift (>> 2)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
