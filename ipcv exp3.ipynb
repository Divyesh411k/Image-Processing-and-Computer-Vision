```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import fashion_mnist
```

```python
(train_X, train_y), (test_X, test_y) = fashion_mnist.load_data()

plt.figure(figsize=(12, 12))
for i in range(9):
  plt.subplot(3,3,i+1)
  plt.imshow(train_X[i])
plt.show()
```

```python
def contrast_stretching(image):
    min = np.min(image)
    max = np.max(image)
    contrast_stretched_img = (image - min) / (max - min) * 255
    return  contrast_stretched_img

num_images = 9
for i in range(9):
  plt.subplot(3,3,i+1)
  plt.imshow(train_X[i])
  plt.title("Original image")
plt.show()

for i in range(num_images):
    contrast_stretched_img =contrast_stretching(train_X[i])
    plt.subplot(3, 3, i + 1 )
    plt.imshow(contrast_stretched_img)
    plt.title("Contrast stretch image")
plt.show()
```

```python
def log_transformation(image):
  c = 255 / np.log(1 + np.max(image))
  log_image = c * (np.log(image + 1))
  log_image = np.array(log_image, dtype = np.uint8)
  return log_image

num_images = 9

for i in range(9):
  plt.subplot(3,3,i+1)
  plt.imshow(train_X[i])
  plt.title("Original image")
plt.show()

for i in range(num_images):
    log_image =log_transformation(train_X[i])
    plt.subplot(3, 3, i + 1 )
    plt.imshow(log_image)
    plt.title("log transform")
plt.show()
```

```python
def power_law_transformation(img,gamavlaue):
  gamma_img = np.array(255*(img/255)**gamavlaue)
  return gamma_img

for i in range(9):
  plt.subplot(3,3,i+1)
  plt.imshow(train_X[i])
  plt.title("Original image")
plt.show()

for i in range(num_images):
    newimg =power_law_transformation(train_X[i],2.2)
    plt.subplot(3, 3, i + 1 )
    plt.imshow(newimg)
    plt.title("Power law")
plt.show()
```

```python
for i in range(num_images):
    newimg =power_law_transformation(train_X[i],5.6)
    plt.subplot(3, 3, i + 1 )
    plt.imshow(newimg)
    plt.title("Power law")

plt.show()
```
